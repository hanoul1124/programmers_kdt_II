## I. Matplotlib 시작하기

### Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리
- `%mathplotlib inline` 통해서 활성화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Case study with Arguments

```python
# 아래 예시에서는 기본적으로 x축이 index가 되고, 주어진 argument list가 y값이 된다
plt.plot([2,4,2,4,2]) # 실제 plotting을 하는 함수
plt.show() # plt 확인하는 명령
```

### Figsize: Figure(도면)의 크기 조절

```python
plt.figure(figsize=(3, 3)) # plotting을 할 도면을 선언
plt.plot([0,1,2,3,4])
plt.show()
```

### 2차함수 그래프 with plot()

```python
# 리스트를 이용해서 1차함수 y=x를 그려보면:
plt.plot([0,1,2,3,4])
plt.show()
```

```python
# numpy.array를 이용해서 함수 그래프 그리기
# y=x^2
x = np.array([1,2,3,4,5]) # 정의역
y = np.array([1,4,9,16,25]) # f(x)

plt.plot(x, y)
plt.show()
```

```python
# np.arnage(a, b, c) c: 0.01
x = np.arange(-10, 10, 0.01)
plt.plot(x, x**2)
plt.show()
```

```python
# x, y 축에 설명 추가하기
x = np.arange(-10, 10, 0.01)

##
plt.xlabel("x value")
plt.ylabel("f(x) value")
##

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축의 범위 설정하기
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

##
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
##

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축에 눈금 설정하기
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]

##
plt.xticks([i for i in range(-5, 6, 1)]) # x축의 눈금 설정, -5, -4, -3 ...
plt.yticks([i for i in range(0, 27, 3)]) # y축의 눈금 설정
##

plt.plot(x, x**2)
plt.show()
```

```python
# 그래프에 title 달기
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 6, 1)]) # x축의 눈금 설정, -5, -4, -3 ...
plt.yticks([i for i in range(0, 27, 3)]) # y축의 눈금 설정

##
plt.title("y = x^2 graph")
##

plt.plot(x, x**2, label="trend")
##
plt.legend() # legend는 plot이 작성된 후 실행되어야 함
##

plt.show()
```

## II. Matplotlib Case Study

### 꺾은선 그래프(Plot)
- 시계열 데이터에서 주로 사용
- `plot()`

```python
x = np.arange(20) # 0~19
y = np.random.randint(0, 20, 20) # 난수를 20번 생성

plt.plot(x, y)

# Extra: y축을 20까지 보이게 하고 싶다면? y축을 "5" 단위로 보이게 하고 싶다면?
# .axis() , .ytics()
plt.axis([0, 20, 0, 20])
plt.yticks([0, 5, 10, 15, 20])

plt.show()
```

### 산점도 (Scatter Plot)
- x, y가 별개의 변수일 때 사용되어, 변수 간 상관관계를 유추할 때 사용
- `.scatter()`

```python
plt.scatter(x, y)
plt.show()
```

### 박스 그림(Box Plot)
- 단일 변수의 전반적인 분포를 확인할 때 자주 사용
- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)
- `.boxplot()`

```python
plt.boxplot((x, y))

# Extra : title 설정
plt.title("Box Plot of x, y")
plt.show()
```

### 막대 그래프(Bar Plot)
- 범주형 데이터에서 자주 사용.
- 범주형 데이터의 값과 그 값의 크기를 직사각형으로 나타낸 그림
- 막대그래프와 히스토그램 모두 빈도의 표현에 적합
- `.bar()`

```python
plt.bar(x, y)
# xticks를 올바르게 처리
plt.xticks(np.arange(0, 20, 1))
plt.show()
```

```python
# cf) Histogram
# .hist()
# 도수분포를 직사각형의 막대 형태로 나타낸다.
# "계급"으로 나타낸 것이 특징 : 0, 1, 2가 아니라 0~2까지의 "범주형" 데이터로 구성 후 그림을 그림
plt.hist(y, bins=np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))
plt.show()
```

### 원형 그래프(Pie Chart)
- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 **비율** 확인에 용이하다
- `.pie()`

```python
z = [100, 200, 300, 400]
plt.pie(z, labels=['one', 'two', 'three', 'four'])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

### Matplotlib를 기반으로 더 다양한 시각화 방법을 제공하는 라이브러리
- 커널밀도그림
- 카운트그림

- 스트립그림
- 히트맵

### Seaborn Import하기

```python
import seaborn as sns
```

### 커널밀도그림(Kernel Density Plot)
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- `sns.kdeplot()`

```python
# in Histogram
x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins=x)
plt.show()
```

```python
# kdeplot
sns.kdeplot(y, shade=True)
plt.show()
```

### 카운트그림 (Count Plot)
- 범주형 column의 빈도수를 시각화 -> Gorupby 후의 도수를 하는 것과 동일한 효과
- `sns.countplot()`

```python
vote_df = pd.DataFrame({"name": ['Andy', 'Bob', 'Cat'], "vote": [True, True, False]})
vote_df
```

```python
# in matplotlib barplot
vote_count = vote_df.groupby('vote').count()
vote_count
```

```python
plt.bar(x=[False, True], height=vote_count['name'])
plt.show()
```

```python
# sns의 countplot
# groupby 작업 없이 원 데이터를 넣어주면 된다
sns.countplot(x=vote_df['vote'])
plt.show()
```

### 캣그림 (Cat Plot) 
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()`
- cat : concatenate. 복잡한 데이터에 용이

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
s = sns.catplot(x='WHO Region', y='Confirmed', data=covid)
# sns figsize
s.fig.set_size_inches(10, 6)
plt.show()
```

```python
# 위 결과는 catplot이 default로 strip plot 을 사용하기 때문에 나타난 그림
# catplot은 다양한 변수를 그래프로 묶어주는 것이지 고정된 그래프 형태가 아니므로, 다른 그래프를 적용할 수도 있다.

s = sns.catplot(x='WHO Region', y='Confirmed', data=covid, kind='violin')
# sns figsize
s.fig.set_size_inches(10, 6)
plt.show()
```

### 스트립그림 (Strip Plot)
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`

```python
sns.stripplot(x='WHO Region', y='Recovered', data=covid)
plt.show()
```

```python
# cf) swarmplot
# stripplot과 거의 유사하나, 점들을 겹치지 않게 분산시킨다
sns.swarmplot(x='WHO Region', y='Recovered', data=covid)
plt.show()
```

### 히트맵(Heatmap)
- 데이터의 행렬을 색상으로 표현해주는 그래프
- `sns.heatmap()`

```python
# 히트맵 예제
# 데이터를 행렬 형태로 표현. correlation 데이터
covid.corr()
```

```python
sns.heatmap(covid.corr())
plt.show()
```
